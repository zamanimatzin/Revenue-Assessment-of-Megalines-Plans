## Aggregate data per user

Now, the data is already clean. Next, data per user per period will be aggregate in order to have just one record per user per period. It will be used for further analysis. 

Firstly, calculate the number of calls made by each user per month. Then, save the result.

```python
calls_per_user_period = calls.pivot_table(index = ['user_id', 'call_date_month'], values = 'duration', aggfunc = ['sum','count'])
calls_per_user_period.columns = ['total_minutes', 'sum_calls']
calls_per_user_period = calls_per_user_period.reset_index()
calls_per_user_period = calls_per_user_period.rename(columns={'call_date_month': 'month'})
calls_per_user_period
```

Next, calculate the amount of minutes spent by each user per month and save the result

```python
messages_per_month = messages.groupby(['user_id','message_date_month'])['id'].agg('count').reset_index()
messages_per_month = messages_per_month.rename(columns={'message_date_month':'month','id':'sum_messages'})
messages_per_month
```

In addtion, calculate the volume of internet traffic used by each user per month and save the result.

```python
internet_per_month = internet.groupby(['user_id','session_date_month'])['mb_used'].agg('sum').reset_index()
internet_per_month = internet_per_month.rename(columns={'session_date_month':'month','mb_used':'total_internet'})
internet_per_month
```

Finally, aggregate data together into one DataFrame in one record. It would represent what an unique user consumed in a given month

```python
# Merge the data for calls, minutes, messages, internet based on user_id and month
merge_data = calls_per_user_period.merge(messages_per_month,on=['user_id','month'],how='outer').merge(internet_per_month,on=['user_id','month'],how='outer')
merge_data
```

After that, input the plan information with the merge_data so we know the plan used for each user.

```python
# Add the plan information
plan_users = users[['user_id','plan']]
merge_data = merge_data.merge(plan_users,on='user_id')
merge_data
```

```python
# Check the percentage missing value
(merge_data.isna().sum()/merge_data.shape[0]*100).sort_values(ascending=False)
```

```python
# Fix the missing value
# Messages
merge_data ['sum_messages'] = merge_data ['sum_messages'].fillna(0)
# Total minutes 
merge_data ['total_minutes'] = merge_data ['total_minutes'].fillna(0)
# Sum calls
merge_data ['sum_calls'] = merge_data ['sum_calls'].fillna(0)
# Total internet
merge_data ['total_internet'] = merge_data ['total_internet'].fillna(0)
merge_data.isna().sum()
```

There is a missing value, so, it signifies that the user did not carry out any of this action during the month. so the value should be filled with 0

Define a function to calculate the monthly revenue from each user based on plan information. Added the result in the merge data. 

```python
# Calculate the monthly revenue for each user

def monthly_revenue(plan,data,calls,sms):
    
    if plan == 'surf':
        plan_cost = 20
        calc_data = math.ceil((data - 15360)/1024)
        if (calc_data <= 0):
            cost_add_data = 0
        elif (calc_data > 0):
            cost_add_data = calc_data * 10
        
        calc_calls = calls - 500
        if (calc_calls <= 0):
            cost_add_calls = 0
        elif (calc_calls > 0):
            cost_add_calls = calc_calls * 0.03
        
        calc_sms = sms - 50
        if (calc_sms <= 0):
            cost_sms = 0
        elif (calc_sms > 0):
            cost_sms = calc_sms * 0.03 
        
    elif plan == 'ultimate':
        plan_cost = 70 
                                
        calc_data = math.ceil((data - 30720)/1024)
        if (calc_data <= 0):
            cost_add_data = 0
        elif (calc_data > 0):
            cost_add_data = calc_data * 7
                                
        calc_calls = calls - 3000
        if (calc_calls <= 0):
            cost_add_calls = 0
        elif (calc_calls > 0):
            cost_add_calls = calc_calls * 0.01

        calc_sms = sms - 1000
        if (calc_sms <= 0):
            cost_sms = 0
        elif (calc_sms > 0):
            cost_sms = calc_sms * 0.01        
        
    total_cost = plan_cost + cost_add_data + cost_add_calls + cost_sms
    return (total_cost)
```

```python
merge_data['monthly_revenue'] = merge_data.apply(lambda x: monthly_revenue(x.plan, x.total_internet, x.total_minutes, x.sum_messages), axis=1)
merge_data.head(10)
```

```python
# Compare the number of minutes users of each plan require each month. Plot a histogram.
slice_data = merge_data[['plan','month','total_minutes']]
slice_data = slice_data.sort_values('month', key = lambda x : pd.Categorical(x, categories=months, ordered=True))

def minutes_users (month):
    month_query = slice_data.loc[slice_data['month'] == month]
    
    for row in month_query:
        surf_minutes_hist = month_query.query('plan == "surf"')
        ultimate_minutes_hist = month_query.query('plan == "ultimate"')
        
    surf_bar['plan']='surf'
    ultimate_bar['plan']='ultimate'
    
    histogram = pd.concat([surf_minutes_hist,ultimate_minutes_hist])
    fig = plt.subplots(figsize =(5, 3))
    plt.title(month)
    bins = 10
    sns.histplot(data=histogram, x="total_minutes",bins = bins, alpha = 0.2, hue="plan")
    return (plt.show())
```

#### Boxplot

Lets calculate the mean and the variable of the call duration to reason on whether users on the different plans have different behaviours for their calls.

```python
# Calculate the mean and the variance of the monthly call duration
def monthly_call (month,plan):
    a = slice_data.loc[slice_data['month'] == month]
    for row in a:
        b = a.loc[a['plan'] == plan]
        mean = b.total_minutes.mean()
        variance = np.var(b.total_minutes)
        std = np.sqrt(variance)
    d = {'month': [month], 'plan': [plan], 'mean': [mean], 'variance': [variance], 'std': [std]}
    df = pd.DataFrame(data = d)
    return (df)
```

```python
frame_list = []
month_list = slice_data.month.unique()
plans_list = slice_data.plan.unique()

for month in month_list:
    for plan in plans_list:
        df = monthly_call(month,plan)
        frame_list.append(df)
        
month_call = pd.concat(frame_list).reset_index()
month_call
```

```python
# Plot a boxplot to visualize the distribution of the monthly call duration
fig = plt.subplots(figsize =(20, 10))
sns.boxplot(x='month',y='total_minutes',data=slice_data,hue='plan')
plt.axhline(y = 500, color = 'r', linestyle = '--',label = "limit surf")
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper center')
plt.show()
```

 **Overview**
 
In Boxplot, both plans maximum call majority around 800 to 1000 minutes. For the surf plan, the IQR average is below than 600, only 100 minutes more than the limit. For the ultimate plan, the IQR and the maximum show almost identical with surf plan. All users in ultimate plan doesn’t even reach from the limit call provided. 

#### Conclusion

In conclusion, the user between surf plan and ultimate plan shows the same behavior which show almost the same number of total minutes for every month. The average of total minutes are around 400 to 500 minutes. Therefore, there are no difference between plan surf and plan ultimate in terms for call. In term of calls perspective, surf plan gives more profit where the subscriber needs to pay more with different cost rate while the ultimate plan gives the same value per month because the subscriber doesn’t over the limit call provided.

## Test statistical hypotheses

**Hypotheses**

There are two plans: Ultimate and Surf calling plans from the sample from different groups of plans users. Test the hypothesis that average revenue from users of Ultimate and Surf calling plans are same. For instance, maybe the average revenue for the both plans are somewhat giving the same profit. Let the significance level be 0.05 which most common significant level used.

```python
surf_filt = merge_data[merge_data['plan'] == 'surf']
surf_filt_revenue = surf_filt['monthly_revenue'] 
monthly_revenue_surf = list (surf_filt_revenue)

ultimate_filt = merge_data[merge_data['plan'] == 'ultimate']
ultimate_filt_revenue = ultimate_filt['monthly_revenue'] 
monthly_revenue_ultimate = list (ultimate_filt_revenue)
                                    

alpha = 0.05 # critical statistical significance level

# test the hypothesis that the means of the two independent populations are equal
results_plan = st.ttest_ind(monthly_revenue_surf, monthly_revenue_ultimate, equal_var = False) 

# p-value 
print('p-value:', results_plan.pvalue) 

if (results_plan.pvalue < alpha):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```
